## Students Mark Analysis

```python
import numpy as np
```

```python
marks = np.random.randint(10, 101, size=(20,5))
```

```python
marks
```

#### 1) Average by subject and student

```python
# Average for Subject
average_by_subject = np.mean(marks,axis=0)
print("Average By Subject",average_by_subject)
```

```python
# Average for Student
average_by_student = np.mean(marks,axis=1)
print("Average By Student",average_by_student)
```

#### 2) Highest & lowest per subject

```python
# Highest Subject
highest = np.max(marks,axis=0)
print("Highest Marks for all Subject",highest)
```

```python
# Lowest Subject
lowest = np.min(marks,axis=0)
print("Lowest Marks for all Subject",lowest)
```

#### 3) Overall class topper

```python
total_marks = np.sum(marks,axis=1)
print("Total Marks",total_marks)
print("")
class_topper = np.max(total_marks)
print("Overall Class Topper Mark",class_topper)
```

#### 4) Pass count per subject

```python
pass_fail = marks >= 35
pass_count = np.sum(pass_fail,axis=0)
print("Pass count of the Subjects",pass_count)
```

#### 5) Which subject is difficult?

```python
difficult_subject = np.argmin(average_by_subject)
print("Difficult subject is",difficult_subject, "with average mark of",average_by_subject[difficult_subject])
```

#### 6) Ranking students

```python
rank = np.argsort(-total_marks)
print("First Rank of the Class", rank[0])
```
