### Pandas Series

#### Create Series

```python
data = [10,20,30,40]
s = pd.Series(data)
print("Series",s)
```

```python
# Apply Indexing
s = pd.Series(data,[4,5,6,7])
print("Indexing",s)
```

```python
# Apply Indexing(a,b,c,d)
s = pd.Series(data,['a','b','c','d'])
print("Alphabet",s)
```

```python
# Numpy to Series
import numpy as np

print(pd.Series(np.array([100,200,300]),index=[1,2,3]))
```

```python
# Dictionary to Series
print(pd.Series({'first_name':'Manoj','last_name':'G','postal_code':637402}))
```

```python
# Label
print(s['b'])
```

```python
# Addition 
print(s + 2)
```

```python
# Concatenation
s1 = pd.Series([10,20,30,40],index = ['a','b','c','d'])
s2 = pd.Series([100,200,300,400],index = ['a','b','c','d'])
print(s1 + s2)
```

```python
s1 = pd.Series([10,20,30],index = ['a','b','c'])
s2 = pd.Series([100,200,300,400],index = ['a','b','c','d'])
print(s1 + s2)
```

```python
# Position based Access
# Ilocation (iloc)
print(s1.iloc[0])
```

```python
# location (loc)
print(s1.loc['a'])
```

### Pandas DataFrame

```python
data = [['Ramesh','Suresh'],[20,40]]
df = pd.DataFrame (data)
print(df)
```

```python
# Dictionary to DataFrame
data = {
    'Name':['Ramesh','Suresh'],
    'Age':[20,40]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Assign the Index Values
df = pd.DataFrame (data, index = [100,101])
print(df)
```

```python
# Read Csv Syntax
# df = pd_read_csv('filename.csv')
# print(df)
```

```python
# Fetch the Particular Values
print(df['Name'])
```

```python
print(type(df))
```

```python
# Add New Column
df['Salary'] = [50000,400000]
print(df)
```

```python
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
df['Net Salary'] = df['Salary'] + df['Bonus']
print(df)
```

```python
# Remove Column
del df['Net Salary']
print(df)
```
