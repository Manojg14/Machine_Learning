```python
# Subplot Import Function from Matplotlib

import matplotlib.pyplot as plt
```

#### Line Plot (Trend Analysis)

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

#### Bar Plot (Category Comparison)

```python
students = ["A", "B", "C", "D"]
marks = [85, 90, 78, 92]

plt.bar(students, marks)
plt.title("Bar Plot Example")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

#### Histogram

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title("Histogram Example")
plt.show()
```

#### Scatter Plot (Relationship Checker)

```python
x = [1,2,3,4,5,6]
y = [2,4,5,4,5,7]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#### Pie Chart

```python
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

#### Figure Size

```python
plt.figure(figsize=(8,6))
plt.plot([1,2,3],[4,5,6])
plt.show()
```

#### Adding Grid

```python
plt.plot([1,2,3],[4,5,6])
plt.grid(True)
plt.show()
```

#### Customization (Color, Marker, Line Style)

```python
x = [5, 7, 8, 7, 2, 17]
y = [99, 86, 87, 88, 100, 86]
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title("Customized Plot")
plt.show()
```
